1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f_x = 10 * x - 5
d_f_x = sp.diff(f_x, x)

print("Похідна лінійної функції 10x-5:", d_f_x)
```

```python
f_func = sp.lambdify(x, f_x, 'numpy')
df_func = sp.lambdify(x, d_f_x, 'numpy')

x_vals = np.linspace(0, 10, 100)
y_vals = f_func(x_vals)
y_df_vals = df_func(x_vals) * np.ones_like(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = 10x - 5', color='blue')
plt.plot(x_vals, y_df_vals, label="df(x) = 10", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("Кількість годин навчання (x)")
plt.ylabel("Оцінка / Зміна оцінки")
plt.legend()
plt.grid(True)
plt.show()
```

**Висновок:**

Так як функція лінійна, то похідна від неї є константою. Тобто функція маю постійну швидкість зміни без прискорень.

З кожною додатковою годиною навчання кількість балів збільшується на 10.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin = sp.sin(x)
f_ln = sp.ln(x)


df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)


f_sin_func = sp.lambdify(x, f_sin, 'numpy')
df_sin_func = sp.lambdify(x, df_sin, 'numpy')

f_ln_func = sp.lambdify(x, f_ln, 'numpy')
df_ln_func = sp.lambdify(x, df_ln, 'numpy')


x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 100)
x_ln = np.linspace(0.1, 10, 100)


y_sin = f_sin_func(x_sin)
y_sin_deriv = df_sin_func(x_sin)

y_ln = f_ln_func(x_ln)
y_ln_deriv = df_ln_func(x_ln)


fig, axs = plt.subplots(1, 2, figsize=(14, 5))

axs[0].plot(x_sin, y_sin, label='f(x) = sin(x)', color='blue')
axs[0].plot(x_sin, y_sin_deriv, label=f"f'(x) = {df_sin}", color='orange', linestyle='--')
axs[0].set_title("Функція sin(x) та її похідна")
axs[0].legend()
axs[0].grid(True)

axs[1].plot(x_ln, y_ln, label='f(x) = ln(x)', color='green')
axs[1].plot(x_ln, y_ln_deriv, label=f"f'(x) = {df_ln}", color='red', linestyle='--')
axs[1].set_title("Функція ln(x) та її похідна")
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f_sqr_x = x**2 - 5*x + 1
df_sqr_x = sp.diff(f_sqr_x, x)


f_sqr_x_func = sp.lambdify(x, f_sqr_x, 'numpy')
df_sqr_x_func = sp.lambdify(x, df_sqr_x, 'numpy')


x_sqr = np.linspace(0, 100, 100)
y_sqr = f_sqr_x_func(x_sqr)
y_sqr_deriv = df_sqr_x_func(x_sqr)


plt.figure(figsize=(10, 5))
plt.plot(x_sqr, y_sqr, label=f'f(x) = {f_sqr_x}', color='blue')
plt.plot(x_sqr, y_sqr_deriv, label=f"df(x) = {df_sqr_x}", color='green', linestyle='--')

plt.title("Графік функції та її похідної")
plt.xlabel("Кількість вироблених одиниць (x)")
plt.ylabel("Витрати компанії")
plt.legend()
plt.grid(True)
plt.show()
```
